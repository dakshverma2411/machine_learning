```python
def costfun(X,y,theta):
    j=0
    l=10
    m,n=X.shape
    for i in range(0,m):
        j=j+((np.matmul(theta.T,X[i].reshape(n,1))-y[i])**2)/(2*m)
    for i in range(1,n):
        j=j+(l*(float(theta.T[0][i])))/(2*m)
    return j
```

```python
def diffarray(X,y,theta):
    m,n=X.shape
    list1=[]
    for j in range(0,n):
        x=0
        for i in range(0,m):
            x=x+((np.matmul(theta.T,X[i].reshape(n,1))-y[i])*X[i][j])/m
        list1.append(x)
#     print("derivative",np.array(list1).reshape(n,1))
    return np.array(list1).reshape(n,1)
```

```python
def gradientdescent(X,y):
    m,n=X.shape
    a=0.01
    l=10
    theta=np.array([0 for i in range(0,n)]).reshape(n,1)
    jprev=costfun(X,y,theta)
    jprev=float(jprev)
    print(jprev)
   
    
    while True:
        temp=(1-((l*a)/m))*theta-a*diffarray(X,y,theta)
        theta=temp.reshape(n,1)
        j=costfun(X,y,theta)
        j=float(j)
        print(j)
        
        if jprev-j<=0.001:
            print("final:",j)
            return theta
        else:
            jprev=j
         
     
            
```
