```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def costfun(theta,X,Y):
    j=0
    m,n=X.shape
    for i in range(0,m):
        if(Y[i]==0):
            j+=((-(math.log(1-hypothesis(np.matmul(theta.T,X[i].reshape(n,1))))))/(2*m))
        else:
            j+=((-(math.log(hypothesis(np.matmul(theta.T,X[i].reshape(n,1))))))/(2*m))
    return j
```

```python
def hypothesis(z):
    ze=-float(z)
    return 1/(1+math.e**(ze))
```

```python
def derivative(theta,X,Y):
    list1=[]
    m,n=X.shape
    for i in range(0,n):
        sum=0
        for j in range(0,m):
            sum+=((hypothesis(np.matmul(theta.T,X[j].reshape(n,1)))-Y[j])*X[j][i])/m
            
        list1.append(sum)
    print("derivative",np.array(list1).reshape(n,1))
    return np.array(list1).reshape(n,1) 
```

```python
def logisticreg(X,Y):
    m,n=X.shape
    theta=np.array([1 for i in range(0,n)]).reshape(n,1)
    while True:
        temp=theta-0.01*derivative(theta,X,Y)
        theta=temp.reshape(n,1)
        count=0
        for i in derivative(theta,X,Y):
            if float(i)>=0.01 or float(i)<=-0.01:
                count+=1
                break
        if count==0:
            return theta
```
