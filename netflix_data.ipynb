```python
dur=df["duration"]
```

```python
list1=[]
for i in dur:
    if "Seasons" in i:
        list1.append(i)
    elif "Season" in i:
        list1.append(i)
    else:
        list1.append("0 Seasons")
    
```

```python
len(list1)
```

```python
list1
```

```python
df["Seasons"]=list1
```

```python
df
```

```python
count=0
for i in df["Seasons"]:
    if int(i[0])>=5:
        count+=1
print(count)
```
